```python
import openai
import os
import csv
import pandas as pd
openai.api_key = os.getenv('OPENAI_API_KEY') 


def find_keywords_fcds(row):
    try:
        query = ' '.join(row)
        response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
        {
            "role": "user",
            "content": f"extract key concepts from the following xml text such that we can assess data science knowledge based on them and put them in a numbered list with no header: {query}"
        }
        ],
        temperature=0.2,
        max_tokens=256,
        top_p=0.1,
        frequency_penalty=0,
        presence_penalty=0
        )
        concepts = response["choices"][0]["message"]["content"]
        concept_list = concepts.split('\n')
        cleaned_concept_list = []
        for item in concept_list:
            cleaned_item = item[item.find('. ') + 2:]
            cleaned_concept_list.append(cleaned_item)

        return cleaned_concept_list
    
    except Exception as e:
        print(f"Error: {e}")
        return 'Error'

def find_keywords_chem(row):
    try:
        query = ' '.join(row)
        response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
        {
            "role": "user",
            "content": f"extract key concepts from the following xml text such that we can assess chemistry knowledge based on them and put them in a numbered list with no header: {query}"
        }
        ],
        temperature=0.2,
        max_tokens=256,
        top_p=0.1,
        frequency_penalty=0,
        presence_penalty=0
        )
        concepts = response["choices"][0]["message"]["content"]
        concept_list = concepts.split('\n')
        cleaned_concept_list = []
        for item in concept_list:
            cleaned_item = item[item.find('. ') + 2:]
            cleaned_concept_list.append(cleaned_item)
        print('done')
        return cleaned_concept_list
    
    except Exception as e:
        print(f"Error: {e}")
        return 'Error'



def fill_empty(row):
    return ''
```
